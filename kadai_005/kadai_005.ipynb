```python
import numpy as np
import matplotlib.pyplot as plt

# グラフを描画する準備
fig, axes = plt.subplots(2, 3, tight_layout=True)
# figは描画領域全体を指す。
# axesは個々のグラフ領域を指す。
# subplots()でgifの中にaxesを当てはめるイメージ
# 第一引数は行数、第二引数は列数、第三引数はサブプロットのレイアウトをある程度自動調整する設定。

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)


x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(x1 * 2)

axes[0, 1].plot(x1, y1, linewidth=2.0)

axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10, 1),
               ylim=(-1, 5), yticks=np.arange(-1, 6, 1))


x3 = np.random.normal(15, 5, 2000)

axes[0, 2].hist(x3, bins = 30)
axes[0, 2].set(xlim=(-10, 40))


np.random.seed(3)

x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))

axes[1, 0].scatter(x4, y4)


x5 = np.array([30, 40, 30])
label = ["dig", "cat", "rabit"]

axes[1, 1].pie(x5, labels=label)
axes[1, 1].set_title("animals I like")


hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]

axes[1, 2].boxplot(hertz)
axes[1, 2].set_title("helz")

plt.show()
```
