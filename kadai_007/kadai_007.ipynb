```python
def tax(x):
  return x * 1.10

df['単価'].apply(tax)

# applyメソッドは、関数を引数に渡してデータを加工するためのもの
# 尚このような簡単な例であれば、df['単価'] * 1.10 でも同様の結果になる。
```

```python
type(df['発注日'].loc[0])
# dfの発注日列のインデックス０行目のデータ型を調べる
```

```python
df['発注日'] = pd.to_datetime(df['発注日'])

# pd.to_datetime()   引数(dfの発注日列全て）のデータ型をデータタイム型へ変更し、それを代入しなおす。
```

```python
type(pd.to_datetime(df['発注日']).loc[1])

# 今回は時間が入っていない為タイムスタンプ型になる。
```

```python
df.agg({'発注日': ['max', 'min']})
# 発注日カラムの最新と最古を抽出

# df.agg()   カラムごとに集計処理をする関数
# {'発注日': ['max', 'min']}   発注日カラムの最新、最古を取得
# つまり発注日カラムの最新と最古を取得し、それを集計するという流れ
```

```python
tax_series = df['単価'].apply(tax)
# dfの単価カラムに、1.1を掛けるtax関数を適用してデータを加工。それを新たな変数に代入

tax_series.name = "単価（税込）"
pd.concat([df, tax_series], axis=1)
# concat(元のデータフレーム, 結合したいシリーズorデータフレーム)
# つまり、元のdfに新たなtax_seriesというシリーズを追加するということ。
# axis=1は、Row方向（行方向に結合する設定）（Columns方向（列方向）に結合する場合はaxis=0）
```

```python
# 他のcsvデータをマージする準備
category_df = pd.read_csv('category.csv')
category_df
```

```python
# pandas.merge(分析対象のデータフレーム, マージしたいデータフレーム[['一致させる要素の列名', '付随する列の名前']], how='inner', on='一致させる要素の列名')

df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df

# マージとは、元のデータと新しいデータの差分をみて更新するという意味。
# how='innerとは、共通する商品番号のみを残すという意味。（もし片方にしか無い商品番号があれば、それはマージ後には削除される。）
# dfに対して、category_dfの「商品番号・カテゴリー」カラムを結合する、但しon商品番号をキーにして融合するというイメージ。
```
